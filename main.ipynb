```python
import pandas as pd
import requests
import json
from collections import defaultdict
```

```python
from acess_token import my_token as token
```

```python
url = "https://api.vk.com/method/groups.getById"
params = {
    'group_ids': "atl_acidhouze",
    "access_token": token,
    'v': '5.89',
    'fields': 'description, members_count'
}
res = requests.get(url, params=params)
res_dict = json.loads(res.content)
group = res_dict['response'][0]
group_name = group['name']
group_members = group['members_count']
group_description = group['description']
print(f'Название группы: {group_name}. Число участников: {group_members}. Описание: {group_description}')
```

```python
url = "https://api.vk.com/method/groups.getMembers"
params = {
    'group_id': "atl_acidhouze",
    "access_token": token,
    'v': '5.89',
    'fields': 'city, sex, universities'
}
res = requests.get(url, params=params)
res_dict = json.loads(res.content)
res_dict = res_dict['response']
res_dict
```

```python
cities = defaultdict(lambda: 0)
for member in res_dict['items']:
    if 'city' in member.keys():
        city_name = member['city']['title']
        cities[city_name] += 1
max_cnt = 0
most_popular_city = ''
for k, v in cities.items():
    if v > max_cnt:
        max_cnt = v
        most_popular_city = k
        
most_popular_city, max_cnt
```

## Анализ статистики группы показывает, что наибольшее количество участников проживает в городе Санкт-Петербург

```python
sex = defaultdict(lambda: 0)
for member in res_dict['items']:
    if 'sex' in member.keys():
        sex[member['sex']] += 1

sex
```

## Проанализировав половую принадлежность участников группы, можно отметить, что мужчин приблизительно в 3 раза больше, чем женщин


```python
names = defaultdict(lambda: 0)
for member in res_dict['items']:
        names[member['first_name'].lower()] += 1

max_cnt = 0
most_popular_name = ''

for k, v in names.items():
    if v > max_cnt:
        max_cnt = v
        most_popular_name = k

most_popular_name, max_cnt
```
